## Question 3 — Arithmetic, Comparison, Logical, and Membership Operators  *(13 points)*

Use the operator families from the lecture to analyze a student's course results.

### What to do
1. Create these variables:
   - `earned_points = 43`
   - `total_points = 50`
   - `attendance = 0.85`
2. **Arithmetic:** Compute `percentage = earned_points / total_points * 100` and print it. Also print `earned_points // 10` (floor division) and `earned_points % 10` (remainder), each labeled so the output is readable.
3. **Comparison and logical:** Compute and print these booleans:
   - `passed = percentage >= 70 and attendance >= 0.80`
   - `honor  = percentage >= 90 or attendance >= 0.95`
   - `needs_review = not passed`
4. **Assignment operator:** The student earns 5 bonus points. Add them to `earned_points` using `+=`, then print the updated value.
5. **Membership:** A list is provided: `topics = ["variables", "strings", "operators"]`. Print the result of `"strings" in topics` and `"loops" not in topics`.
6. Add a **comment** at the end answering: in step 3, why is `passed` `True` or `False` for these values? (One short sentence.)

### Hints
- Label outputs by passing text first: `print("Percentage:", percentage)`.
- `and` requires **both** sides to be `True`; `or` requires **at least one**.
- `points += 5` is shorthand for `points = points + 5`.

````python
topics = ["variables", "strings", "operators"]

earned_points = 43
total_points = 50
attendance = 0.85
percentage = earned_points / total_points * 100
print("Percentage:", percentage)
print("Earned Points // 10: ", earned_points // 10)
print("Earned point % 10: ", earned_points % 10)
passed = percentage >= 70 and attendance >= 0.80
honor = percentage >= 90 or attendance >= 0.95
needs_review = not passed
print("Passed: ", passed)
print("Honor: ", honor)
print("Needs Review: ", needs_review)
earned_points += 5
print("Updated earned points: ", earned_points)
print('"strings" in topics: ', "strings" in topics)
print('"loops" not in topics:', "loops" not in topics)
# Passed is true because both conditions are met passed and honor.

````

## Question 4 — Random Numbers  *(12 points)*

Use Python's `random` module to build a tiny "lucky student" program.

### What to do
1. Import the `random` module.
2. Set the seed to your favorite whole number with `random.seed(...)` so your results are **reproducible** (running the cell twice gives the same output).
3. Generate and print, with labels:
   - a random float between 0 and 1 using `random.random()`
   - a random integer between 1 and 100 using `random.randint(1, 100)`
4. A list of students is provided in the code cell. Using that list:
   - print **one** randomly chosen student with `random.choice(...)`
   - print **two different** randomly chosen students with `random.sample(..., 2)`
   - shuffle the whole list with `random.shuffle(...)` and print the shuffled list
5. Add a **comment** answering: what does `random.seed(...)` do, and why is it useful? (One or two short sentences.)

### Hints
- `random.shuffle(students)` changes the list **in place** — it returns `None`, so shuffle first, then print the list on the next line.
- Run the cell twice: with a seed set at the top, the output should be identical both times.

````python
import random

random.seed(10)
print("Random float (0-1):", random.random())
print("Random integer (1-100):", random.randint(1, 100))
students = ["Alice", "Ben", "Carla", "David", "Eva", "Frank"]
print("Random student:", random.choice(students))
print("Two random students:", random.sample(students, 2))
random.shuffle(students)
print("Shuffled students:", students)
# random seed sets starting point for random number generator.
# It is useful because makes results interchangeable producing
# random numbers every time.



````
